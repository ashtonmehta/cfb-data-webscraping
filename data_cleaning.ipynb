```python
import pandas as pd
```

```python
df = pd.read_csv('DraftData.csv')
df.head()
```

```python
df[['College Stats URL']].info()
```

```python
df.dropna(subset=['College Stats URL'], inplace=True)
df[['College Stats URL']].info()
```

```python
df['Player'].value_counts()
```

```python
df[df['Player']  == 'Byron Murphy']
```

```python
df = df[['Player', 'Pos', 'College Stats URL']]
```

```python
df.head()
```

```python
df_passing_standard = df[df['Pos'] == 'QB']
df_rushing_standard = df[df['Pos'] == 'RB']
df_receiving_standard = df[(df['Pos'] == 'WR') | (df['Pos'] == 'TE')]
df_defense_standard = df[~((df['Pos'] == 'QB') | (df['Pos'] == 'RB') | (df['Pos'] == 'WR') | (df['Pos'] == 'TE'))]
```

```python
df_defense_standard['Pos'].value_counts()
```

```python
df_passing_standard.to_csv('passing_standard.csv', index=False)
df_rushing_standard.to_csv('rushing_standard.csv', index=False)
df_receiving_standard.to_csv('receiving_standard.csv', index=False)
df_defense_standard.to_csv('defense_standard.csv', index=False)
```
